```python
intents = []
for i, j in components.intent_mapping.items():
    for item in j['intents']:
        intents.append(item)
intents = list(set(intents))

data_types = list(components.intent_mapping.keys())
models = list(components.model_mapping.keys())
datasets = list(components.dataset_mapping.keys())
explainers = list(components.exp_mapping.keys())
```

```python
num = 1
for dt in data_types:
    for dataset in datasets:
        #supported datatypes of the dataset
        supp_dt_dataset = components.dataset_mapping[dataset]['data_type']
        if dt in supp_dt_dataset:
            for intent in components.intent_mapping[dt]['intents']:
                for model in models:
                    if dt in components.model_mapping[model]['task']:
                        for explainer in explainers:
                            if dt in components.exp_mapping[explainer]['data_type']:
                                for pres in components.exp_mapping[explainer]['presentation']:
                                    print(num, dataset, model, intent, explainer, pres)
                                    num += 1
            
```
